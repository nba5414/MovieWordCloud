```python
# Getting the files into a single string
fansfile=open('fansfinal.txt', 'r')
fanstext=fansfile.read()

criticfile=open('criticsfinal.txt', 'r')
critictext=criticfile.read()

fanstext=fanstext.lower()
critctext=critictext.lower()

```

```python
# The text file with the critic reviews had some symbols on it.  This splits them up from the rest of the words so the 
# wordcloud doesn't include them
critictext=critictext.split('&')
critictext=' '.join(critictext)
```

```python
# Function to remove stopwords
def remove_stopwords(text):
    text=text.split(' ')
    newtext=[]
    for i in range(len(text)):
        if text[i] in set(stopwords.words('english')):
            continue
        else:
            newtext.append(text[i])
    
    newtext=' '.join(newtext)
    return newtext
    
fanstextc=remove_stopwords(fanstext)
critictextc=remove_stopwords(critictext)
```

```python
#Function to remove words that shouldn't be included
bad_words=['apos', 'two', 'kick', 'ass', 'bitch', 'ways', 'watch', 'lot', 'gonna', 'much', 'clear', 'guys','guess', 'still', '&amp'
          , 'bitch']

def remove_bad_words(text):
    text=text.split(' ')
    newtext=[]
    for i in range(len(text)):
        if text[i] in set(bad_words):
            continue
        else:
            newtext.append(text[i])
    
    newtext=' '.join(newtext)
    return newtext

fanstextc=remove_bad_words(fanstextc)
critictextc=remove_bad_words(critictextc)

```
